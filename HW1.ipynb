```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
N = 100000
mu = 3
std = .01
noise = np.random.normal(mu,std,N)
lower = 2.95
upper = 3.05
bins = 50

normalization = (upper-lower)/bins*N
x = np.linspace(lower, upper, 100)

plt.hist(noise,bins,(lower,upper))
plt.plot(x, stats.norm.pdf(x,mu,std)*normalization, color='red')
plt.show()

plt.hist(noise,bins,(lower,upper))
plt.plot(x, stats.norm.pdf(x,mu,std)*normalization, color='red')
plt.yscale("log")
plt.show()
```
